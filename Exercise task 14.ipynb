```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer
import torch

# Load the model and tokenizer
model_name = "deepset/xlm-roberta-large-squad2"
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Define question-context pairs directly in context
pairs = [
    {
        "context": "Eiffel-torni sijaitsee Pariisissa, Ranskassa.",
        "question": "Missä Eiffel-torni sijaitsee?",
        "expected_answer": "Pariisissa, Ranskassa"
    },
    {
        "context": "Vesi kiehuu 100 celsiusasteessa.",
        "question": "Missä lämpötilassa vesi kiehuu?",
        "expected_answer": "100 celsiusasteessa"
    }
]

# Evaluate predictions
for pair in pairs:
    # Encode inputs
    inputs = tokenizer(pair['question'], pair['context'], return_tensors='pt')

    # Get model predictions
    with torch.no_grad():
        outputs = model(**inputs)

    # Find start and end positions of the answer
    answer_start = torch.argmax(outputs.start_logits)  # get the most likely start token index
    answer_end = torch.argmax(outputs.end_logits) + 1  # get the most likely end token index

    # Convert tokens to string answer
    predicted_answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(inputs['input_ids'][0][answer_start:answer_end]))

    # Print the result
    print(f"Question: {pair['question']}")
    print(f"Predicted Answer: {predicted_answer}")
    print(f"Expected Answer: {pair['expected_answer']}\n")
```

```python
from transformers import pipeline

# Load the model
qa_pipeline = pipeline("question-answering", model="deepset/xlm-roberta-large-squad2")

# Define question-context pairs
examples = [
    {
        "context": "The Eiffel Tower is a wrought-iron lattice tower on the Champ de Mars in Paris, France.",
        "question": "Where is the Eiffel Tower located?"
    },
    {
        "context": "La Torre Eiffel es una torre de celosía de hierro forjado en el Campo de Marte en París, Francia.",
        "question": "¿Dónde se encuentra la Torre Eiffel?"
    }
]

# Run the model on each example
for example in examples:
    result = qa_pipeline(question=example["question"], context=example["context"])
    print(f"Question: {example['question']}")
    print(f"Answer: {result['answer']}\n")

# Optional: Evaluate using TurkuNLP/squad_v2_fi
# Uncomment the following lines if you want to run the evaluation
# from datasets import load_dataset
# import evaluate

# dataset = load_dataset("TurkuNLP/squad_v2_fi", split="validation")
# metric = evaluate.load("squad_v2")

# predictions = []
# references = []

# for example in dataset:
#     result = qa_pipeline(question=example["question"], context=example["context"])
#     predictions.append({"id": example["id"], "prediction_text": result["answer"], "no_answer_probability": result["score"]})
#     references.append({"id": example["id"], "answers": example["answers"]})

# results = metric.compute(predictions=predictions, references=references)
# print(results)
```

```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline
model_name = "deepset/xlm-roberta-large-squad2"
nlp = pipeline('question-answering', model=model_name, tokenizer=model_name)

QA_input = {
    'question': 'Why is model conversion important?',
    'context': 'The option to convert models between FARM and transformers gives freedom to the user and lets people easily switch between frameworks.'}

#Test
res = nlp(QA_input)
print('Why is model conversion important?')
print(res['answer'])
```
